```python
NL_East_Teams = ['Phillies', 'Braves', 'Nationals', 'Marlins', 'Mets']
```

```python
pitcher_url = 'http://www.espn.com/mlb/stats/pitching/_/seasontype/2/count/41/qualified/true/order/false'
pitch_page = requests.get(pitcher_url)

soup = BeautifulSoup(pitch_page.text, 'html.parser')
```

```python
header = soup.find('tr', attrs={'class': 'colhead'})
```

```python
columns = [col.get_text() for col in header.find_all('td')]
```

```python
pitch_final_df = pd.DataFrame(columns=columns)
pitch_final_df
```

```python
for i in range(1,244, 40):
    url ='http://www.espn.com/mlb/stats/pitching/_/seasontype/2/count/{}/qualified/false/order/false'.format(i)
    #url1 = ''
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')

    players = soup.find_all('tr', attrs={'class':re.compile('row player-10-')})
    for player in players:
        # Get Stat for each player
        stats = [stat.get_text() for stat in player.find_all('td')]
        # Create temp dataframe to add later
        temp_df1 = pd.DataFrame(stats).transpose()
        temp_df1.columns = columns
        # Add Data to final_df
        pitch_final_df = pd.concat([pitch_final_df, temp_df1], ignore_index=True)
pitch_final_df
```

```python
merged = pd.merge(fa_data[['PLAYER', "AGE"]], pitch_final_df, how='left', on="PLAYER")
```
